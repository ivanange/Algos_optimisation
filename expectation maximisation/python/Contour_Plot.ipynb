```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-10.0, 10.0, 100)
x2 = np.linspace(-10.0, 10.0, 100)
```

```python
fig , ax = plt.subplots()
ax.scatter(x1, x2);
```

```python
X1, X2 = np.meshgrid(x1, x2)
Y = np.sqrt(np.square(X1) + np.square(X2))
```

```python
cm = plt.cm.get_cmap('viridis')
fig, ax = plt.subplots()
ax.scatter(X1, X2, c=Y, cmap=cm)
```

```python

fig, ax = plt.subplots()
cp = ax.contour(X1, X2, Y)
ax.clabel(cp, inline=1, fontsize=10)
ax.set(title = "Plot with matplotlib",
      xlabel = "X-axis",
      ylabel = "Y-axis",
      );
```
